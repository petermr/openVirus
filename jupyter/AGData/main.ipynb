```python
#!pip install opencage

from opencage.geocoder import OpenCageGeocode

### Need to get personal api key from https://opencagedata.com
key = "b4857565fb3740e5b5cabe83e10e6bbb"  ### put your api key here
# get api key from:  https://opencagedata.com

geocoder = OpenCageGeocode(key)
```

```python
list_lat = []   # create empty lists

list_long = []


	
for index, row in country_drug.iterrows(): # iterate over rows in dataframe

    City = row['CountryName']
    query = str(City)

    results = geocoder.geocode(query)   
    lat = results[0]['geometry']['lat']
    long = results[0]['geometry']['lng']

    list_lat.append(lat)
    list_long.append(long)

	
# create new columns from lists    

country_drug['lat'] = list_lat   

country_drug['lon'] = list_long


list_lat2 = []   # create empty lists

list_long2 = []


	
for index, row in country_disease.iterrows(): # iterate over rows in dataframe

    City = row['CountryName']
    query = str(City)

    results = geocoder.geocode(query)   
    lat = results[0]['geometry']['lat']
    long = results[0]['geometry']['lng']

    list_lat2.append(lat)
    list_long2.append(long)

	
# create new columns from lists    

country_disease['lat'] = list_lat2   

country_disease['lon'] = list_long2
```

```python
#!pip install folium
# Create a world map to show distributions of drugs reported in papers 
import folium
from folium.plugins import MarkerCluster
#empty map
locations = country_drug[['lat', 'lon']]

locationlist = locations.values.tolist()
world_map= folium.Map(tiles="cartodbpositron")
marker_cluster = MarkerCluster().add_to(world_map)
country_drug2=country_drug.drop(['lat','lon'], axis=1)

#for each coordinate, create circlemarker 
for i in range(0, len(locationlist)):
        radius=5
        country_drug1=country_drug2.iloc[i].to_frame()
        country_drug1 = country_drug1[(country_drug1.T != 0).any()]
       
        
        folium.CircleMarker(locationlist[i], radius=radius, popup= country_drug1.to_html(), fill =True).add_to(marker_cluster)
        

#show the map and save it as worldmap.html
world_map
world_map.save('worldmap.html')
```
